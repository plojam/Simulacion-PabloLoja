```python
import math
import random
import matplotlib.pyplot as plt
import numpy as np
```

### Distribucion de cauchy

#### Enunciado

El ministerio del ambiente comenzo un proyecto donde requiere calcular el aumento y la disminucion del caudal del rio Yanuncay ubicado en Cuenca, para este estudio se evalua la cantidad de lluvia presentada diariamente (representada en valores de 0 a 1) por 100 dias.


```python
iteraciones = 100

#numeros
random.seed(496001)
r = []
[r.append(random.random())for i in range(iteraciones)]

r = sorted(r)

#normal
densidad = []

for i in range(iteraciones):
    #tan pi u 
    resultado = 1/(math.pi*(1+r[i]*r[i]))
    densidad.append(resultado)



#inversa
inversa = []

for i in range(iteraciones):
    #tan pi u 
    #resultado = np.abs(math.tan(math.pi*(r[i]-0.5)))
    resultado = math.tan(math.pi*(r[i]-0.5))
    inversa.append(resultado)

```

```python
print('i \t xi \t resultado')
for i in range(iteraciones):
    print(str(i+1) + ' \t ' + str(r[i]) + ' \t ' + str(inversa[i]))

```

```python
plt.title('Grafico Función')
plt.xlabel('Valores')
plt.ylabel('Resultado')
plt.plot(r,densidad)
plt.show()
```

```python
plt.title('Grafico Función')
plt.xlabel('Valores')
plt.ylabel('Resultado')
plt.plot(r,inversa)
plt.show()
```
