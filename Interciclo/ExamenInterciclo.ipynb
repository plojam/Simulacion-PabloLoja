```python
import math
import random
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Cuadrados medios
def funCuadradosMedios(xo,d,iteraciones):
    final = []
    xn = xo
    # print("It. \t Xn \t Ui \t Rn")
    for i in range(iteraciones):
        xnn = xn**2
        txnn = str(xnn).zfill(8)
        tam = len(txnn)
        ui = int(txnn[int(tam/2-d/2):int(tam/2+d/2)])
        rn = ui / (int('9'*d)+1)
    #   print(str(i) + "\t" + str(xn) + "\t"+ str(ui)+"\t"+str(rn))
        final.append(rn)
        xn = ui

    return final
```

```python
#Congruencia lineal
def funCongruenciaLineal(xo,a,b,m,iteraciones):
    final = []
    xn = xo
    # print("It. \t Xn \t Ui")
    for i in range(iteraciones):
        xnn = (a*xn+b) % m
        ui = xnn/m
        # print(str(i) + "\t"+ str(xnn)+"\t"+str(ui))
        final.append(ui)
        xn = xnn
    return final
```

```python
#tausworthe
def binario_a_decimal(numero_binario):
	numero_decimal = 0 

	for posicion, digito_string in enumerate(numero_binario[::-1]):
		numero_decimal += int(digito_string) * 2 ** posicion

	return numero_decimal
def funTausworthe(r,q,l,iteraciones):
    bits =[]

    #ingresar los valores [0,1] personalmente para el vector inicial
    # for x in range(0,q):
    #     bit = input('Ingrese un valor binario para el vector inicial de longitud '+str(q))
    #     bits.append(int(bit))

    #ooo, llenar directamente el vector inicial con 1
    for x in range(0,q):
        bits.append(1)

    # cantidad = (2**q) - 1 - q
    cantidad = iteraciones*l

    for i in range(q+1,cantidad+1):
        if bits[(i-r)-1] == bits[(i-q)-1]:
            bits.append(0)
        else:
            bits.append(1)

    vector = ''
    for x in range(0,len(bits)):
        vector = vector + str(bits[x])
        if (x+1) % q == 0:
            vector += ' '

    # print(vector)
    nGrupos = math.floor(len(bits)/l)
    # print('Se obtienen ',nGrupos, ' numeros decimales conformados de ', l, ' bits')
    # print('Sobran ', len(bits)-(l*nGrupos), ' bits no utilizados')

    grupos=[bits[i:i + l] for i in range(0, len(bits), l)]

    decimales = []

    for x in grupos:
        binario = ''
        for b in x:
            binario += str(b)
        decimales.append(binario_a_decimal(binario))

    dividido = ''
    divisor = 2**l
    final = []
    for x in decimales:
        dividido += str(x) + '/' + str(divisor) + ",  "
        final.append(x/divisor)

    # print(dividido)
    # print('Final: ',final)
    return final
```

```python
#Random python
def funRandomPython(seed,iteraciones):
    final = []
    random.seed(seed)
    [final.append(random.random())for i in range(iteraciones)]
    return final
```

```python
def diferentes(numero):
    datos = {}

    for x in numero:
        datos[x] = 0
    
    for x in numero:
        datos[x] = datos[x]+1
    
    sen = True
    for x in datos:
        if datos[x] > 1:
            sen = False
            break


    if(sen==True):
        return True
    else:
        return False

def par(numero):
    datos = {}

    for x in numero:
        datos[x] = 0
    
    for x in numero:
        datos[x] = datos[x]+1
    
    sen = 0

    for x in datos:
        if datos[x] == 2:
            sen = sen + 1

    sen3 = False
    for x in datos:
        if datos[x] == 3:
            sen3 = True
            break

    if sen==1 and sen3==False:
        return True
    else:
        return False

def tercia(numero):
    datos = {}

    for x in numero:
        datos[x] = 0
    
    for x in numero:
        datos[x] = datos[x]+1
    
    sen3 = False
    for x in datos:
        if datos[x] == 3:
            sen3 = True
            break

    sen2 = False
    for x in datos:
        if datos[x] == 2:
            sen2 = True
            break

    if(sen2==False and sen3==True):
        return True
    else:
        return False

def doblePar(numero):
    datos = {}

    for x in numero:
        datos[x] = 0
    
    for x in numero:
        datos[x] = datos[x]+1
    
    sen = 0

    for x in datos:
        if datos[x] == 2:
            sen = sen + 1

    if sen==2:
        return True
    else:
        return False

def terciaPar(numero):
    datos = {}

    for x in numero:
        datos[x] = 0
    
    for x in numero:
        datos[x] = datos[x]+1
    
    sen3 = False
    for x in datos:
        if datos[x] == 3:
            sen3 = True
            break

    sen2 = False
    for x in datos:
        if datos[x] == 2:
            sen2 = True
            break
    

    if(sen2==True and sen3==True):
        return True
    else:
        return False
    

def poker(numero):
    datos = {}

    for x in numero:
        datos[x] = 0
    
    for x in numero:
        datos[x] = datos[x]+1
    
    sen = False
    for x in datos:
        if datos[x] == 4:
            sen = True
            break
    
    return sen
    

def quintilla(numero):
    datos = {}

    for x in numero:
        datos[x] = 0
    
    for x in numero:
        datos[x] = datos[x]+1
    
    sen = False
    for x in datos:
        if datos[x] == 5:
            sen = True
            break
    return sen

#poker
def pruebaPoker(serie, x2):
    q = 0
    po = 0
    tp = 0
    dp = 0
    t = 0
    p1 = 0
    d = 0

    for x in serie:
        nS = str(x)
        

        if len(nS)>=7:
            numStr = nS[2:7]
        else:
            numStr = nS[2:]

        # if(len(numStr)!=5):
        #     print(numStr)

        # if(len(numStr)>1 and len(numStr)<5):
        #     while len(numStr)<5:
        #         numStr = numStr + '0'

        if(len(numStr)==1):
            d += 1
        else:

        
            if(quintilla(numStr)==True):
                q += 1
            elif(poker(numStr)==True):
                po += 1
            elif(terciaPar(numStr)==True):
                tp += 1
            elif(doblePar(numStr)==True):
                dp += 1
            elif(tercia(numStr)==True):
                t += 1
            elif(par(numStr)==True):
                p1 += 1
            elif(diferentes(numStr)==True):
                d += 1
    
            

    total = q + po + tp + dp + t + p1 + d
    print('total: ' + str(total))
    eq = 0.0001*total
    epo = 0.0045*total
    etp = 0.009*total
    edp = 0.1080*total
    et = 0.072*total
    ep1 = 0.5040*total
    ed = 0.3024*total

    fq = ((eq-q)**2)/eq
    fpo = ((epo-po)**2)/epo
    ftp = ((etp-tp)**2)/etp
    fdp = ((edp-dp)**2)/edp
    ft = ((et-t)**2)/et
    fp1 = ((ep1-p1)**2)/ep1
    fd = ((ed-d)**2)/ed
    print('Categoria \t O \t E \t Formula')
    print('Quintilla \t |' + str(q) + '\t |' + str(eq) + '\t |' + str(fq))
    print('Poker \t |' + str(po) + '\t |' + str(epo) + '\t |' + str(fpo))
    print('Tercia y par \t |' + str(tp) + '\t |' + str(etp) + '\t |' + str(ftp))
    print('Doble par \t |' + str(dp) + '\t |' + str(edp) + '\t |' + str(fdp))
    print('Tercia \t |' + str(t) + '\t |' + str(et) + '\t |' + str(ft))
    print('Par \t |' + str(p1) + '\t |' + str(ep1) + '\t |' + str(fp1))
    print('Diferente \t |' + str(d) + '\t |' + str(ed) + '\t |' + str(fd))
    # print(po)
    # print(tp)
    # print(dp)
    # print(t)
    # print(p1)
    # print(d)

    suma = fq + fpo + ftp + fdp + ft + fp1 + fd

    print(suma)

    if(suma<=x2):
        print('La serie es independiente')
    else:
        print('La serie no es independiente')

```
