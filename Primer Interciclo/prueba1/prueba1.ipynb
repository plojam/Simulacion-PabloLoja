```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import io
import plotly.graph_objects as go
import math
```

```python
datos = pd.read_excel('datasetmodificado.xlsx',header=None)

datos.head(10)
```

```python
#Generar gráfico de barras de solo las empresas que han mejorado y/o empeorado su ranking con respecto al año anterior.
datos[4][1]

mejoraron = []
empeoraron = []
for x in range(1,len(datos)):
    try:
        if(int(datos[2][x])>0):
            mejoraron.append(x)
        elif(int(datos[2][x])<0):
            empeoraron.append(x)
    except:
        1+1

# print(len(mejoraron))
# print(len(empeoraron))

plt.figure()
barraas = plt.bar(['Mejoraron','Empeoraron'],[len(mejoraron),len(empeoraron)])
plt.title('Rankin')

plt.show()

```

```python
#Generar gráfico de las empresas que tienen panamá papers

tienen = []
notienen = []
for x in range(1,len(datos)):
    try:
        if(int(datos[7][x])>0):
            tienen.append(x)
        else:
            notienen.append(x)
    except:
        1+1

# print(len(mejoraron))
# print(len(empeoraron))

plt.figure()
barraas = plt.bar(['Tienen','No tienen'],[len(tienen),len(notienen)])
plt.title('Panama Papers')

plt.show()
```

```python
#Generar un pie de las empresas que más aportan a la economía del país

#top 8 empresas que mas han colaborado


contribucion = [0,0,0,0,0,0,0,0]
empresas = ['','','','','','','','']

for x in range(1,len(datos)):
    try:
        if(datos[17][x]<=8):
            contribucion[datos[17][x]-1] = datos[20][x]
            empresas[datos[17][x]-1] = datos[4][x]
        # print(datos[20][1])
        # break
        # suma += int(datos[20][x])
    except:
        1+1


plt.figure()

plt.pie(contribucion, labels=empresas, autopct='%.2f %%')
plt.title('8 Empresas que mas contribuyen al Ecuador')

plt.show()
```
